```sage
R = RealField(100)
(R(2).sin().nth_root(6))**10
```

```sage
f(x) = x^2*(sqrt(x^4+16)-x^2)
plot(f,(x,0,2e4))
```

```sage
g = fast_callable(f, vars=[x], domain=R)
plot(g,(x,0,2e4))
```

```sage
u, v = var('u, v')
f_1(u,v) = v*cos(u)
f_2(u,v) = v*sin(u)
parametric_plot3d([f_1,f_2,0],(u,0,pi/2),(v,0.5,1))
```

```sage
g_1 = fast_callable(f_1, vars=[u,v], domain=R)
g_2 = fast_callable(f_2, vars=[u,v], domain=R)
parametric_plot3d([g_1,g_2,0],(u,0,pi/2),(v,0.5,1))
```
